### Import LIbs


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/mmm_dataset.csv")

df["date"] = pd.to_datetime(df["date"])

df = df.sort_values("date")

df.head()
```

### Review structure

```python
print('---------------------------')
print('INFORMATION ABOUT THE DATASET')
print('---------------------------')
print(df.info())
print('---------------------------')
print('STATISTICAL SUMMARY OF THE DATASET')
print('---------------------------')
print(df.describe())
```

### Plots

```python
#Plot sales over time

plt.figure(figsize=(12,6))
plt.plot(df["date"], df["sales"], linestyle='-', color='gray')
plt.title("Sales over time", fontsize=16, color='black', fontweight='bold', pad=20, fontfamily='Arial')
plt.xlabel("Date", fontsize=12, color='black', fontweight='bold', labelpad=10, rotation=0, ha='right', va='center', fontfamily='Arial')
plt.ylabel("Sales", fontsize=12, color='black', fontweight='bold', labelpad=10, rotation=0, ha='right', va='center', fontfamily='Arial') 
plt.xticks(rotation=45, ha='right', fontsize=10, color='black', fontfamily='Arial')
plt.yticks(fontsize=10, color='black', fontfamily='Arial')
plt.tight_layout()
plt.show()
```

```python
# Plot of every channel over time

media_channels = ["tv", "youtube", "facebook", "search"]
colors = ["orange", "red", "blue", "green"]
for channel in media_channels:

    plt.figure(figsize=(12,4))
    plt.plot(df["date"], df[channel], color=colors[media_channels.index(channel)])
    plt.title(channel)
    plt.xlabel("Date", fontsize=12, color='black', fontweight='bold', labelpad=10, rotation=0, ha='right', va='center', fontfamily='Arial')
    plt.ylabel("Spend", fontsize=12, color='black', fontweight='bold', labelpad=10, rotation=0, ha='right', va='center', fontfamily='Arial')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

    plt.show()

```

```python
# Correlation matrix

corr = df.drop(columns=["date"]).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=16, color='black', fontweight='bold', pad=20, fontfamily='Arial')
plt.xticks(rotation=45, ha='right', fontsize=10, color='black', fontfamily='Arial')
plt.yticks(rotation=0, fontsize=10, color='black', fontfamily='Arial')
plt.show()

```

TV and YouTube show the highest correlation with sales (~0.17), suggesting stronger influence compared to Facebook and Search. However, correlations remain low overall, which is expected due to adstock carryover effects, seasonality, and noise. This highlights the need for a proper Marketing Mix Model instead of relying on simple correlations.
